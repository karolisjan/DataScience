```python
import utils
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from neural_network import NeuralNetwork
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss

np.random.seed(42)
```

## Dataset

```python
X, y = datasets.make_moons(200, noise=0.20)

fig = plt.figure(figsize=(10, 10))
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
plt.axis('off')
plt.show()
```

## Logistic Regression

Logist Regression is a linear classification algorithm.

```python
model = LogisticRegression().fit(X, y)

accuracy = accuracy_score(y, model.predict(X)) * 100

utils.plot_decision_boundary(
    title='Logistic Regression\nAccuracy: %.1f%%\nLoss: %.3f' 
        % (accuracy, log_loss(y, model.predict(X))),
    figsize=(10, 10),
    func=lambda x: model.predict(x),
    X=X,
    y=y
)
```

## Neural Network

### 1 hidden layer with 100 units

```python
nn = NeuralNetwork().fit(X, y)

accuracy = accuracy_score(y, nn.predict(X)) * 100

utils.plot_decision_boundary(
    title='Neural Network\nAccuracy: %.1f%%\nLoss: %.3f' 
        % (accuracy, nn.score(X, y)),
    figsize=(10, 10),
    func=lambda x: nn.predict(x),
    X=X,
    y=y
)
```

### 3 hidden layers with 100 units each

```python
nn = NeuralNetwork(
    hidden_layer_sizes=(100, 100, 100,)
).fit(X, y)

accuracy = accuracy_score(y, nn.predict(X)) * 100

utils.plot_decision_boundary(
    title='Neural Network\nAccuracy: %.1f%%\nLoss: %.3f' 
        % (accuracy, nn.score(X, y)),
    figsize=(10, 10),
    func=lambda x: nn.predict(x),
    X=X,
    y=y
)
```
