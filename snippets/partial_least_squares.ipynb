## Linear PLS regression

PLS regression decomposes both $X$ and $Y$ as a product of a common set of orthogonal factors and a set of specific loadings. So, the independent variables are *decomposed* as:

$$
    X = TP^{\prime} + E \quad \text{with} \quad T^{\prime}T = I \text{ and } 
$$

$$
    Y = UQ^{\prime} + F \quad \text{with} \quad U^{\prime}U = I
$$

$Y$ is estimated with:

$$
    \hat{Y} = TBQ^{\prime}
$$

where:
* $X \in \mathcal{R}^{n \times N}$ is a $| \text{ samples }| \times | \text{ predictors } |$ input matrix
* $T$ is a *score* matrix and $P$ is a *loading* matrix; the columns of $T$ are *latent vectors*
* $Y \in \mathcal{R}^{n \times M}$ is a $| \text{ samples }| \times | \text{ responses } |$ output matrix
* $B$ is a diagonal matrix with the *regression weights* as diagonal elements
